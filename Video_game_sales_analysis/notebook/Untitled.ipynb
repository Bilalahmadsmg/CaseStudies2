```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('C:\\Users\\DELL\\Basecamp3\\GLabs_Data_Science_Learn\\Video_game_sales_analysis\\data\\vgsales.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
data['Genre'].value_counts()
```

```python
data['Genre'].value_counts().plot(kind='bar', stacked=True, figsize=(8,5))
plt.xlabel('Genre type')
plt.ylabel('Numbers')
```

## Which year had the most game releases?


```python
data['Year'].value_counts().plot(kind='bar', stacked=True, figsize=(8,5))
```

## What are the top game genres produced for the years(5) with maximum game production?

```python
AA = data.groupby('Genre')['Year'].value_counts().iloc[:5]
AA.plot(kind='bar',stacked=False, figsize=(30,20))
```

```python
plt.figure(figsize=(30,20))
order = data.Year.value_counts().iloc[:5].index
sns.countplot(x='Year', data=data, hue='Genre', order = order)
plt.xticks(size=20,rotation=45)
plt.show()
```
