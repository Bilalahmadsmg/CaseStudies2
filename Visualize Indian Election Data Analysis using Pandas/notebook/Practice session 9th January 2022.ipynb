#### Task 1: Company M & F in bar chart

```python
candidates2009['Candidate_Sex'].value_counts()
```

```python
gender_counts = pd.DataFrame(candidates2009['Candidate_Sex'].value_counts())
gender_counts
```

```python
fig = plt.figure(figsize=(6,4))
plt.bar(height=gender_counts['Candidate_Sex'], x=gender_counts.index)
for num, val in enumerate(gender_counts['Candidate_Sex']):
    plt.text(x=num, y=val+500, s=str(val))
plt.xlabel('Candidate Sex')
plt.ylabel('Count')
plt.ylim(0, 10000)
plt.show
```

#### Task 6: Stacked bar chart to compare seats won by alliances in Gujarat, Madha Pradesh & Maharashtra

```python
states = candidates2009[(candidates2009['Position']==1.0) & (candidates2009['State_name'].isin(['Gujarat', 'Madhya Pradesh', 'Maharashtra']))]
```

```python
states
```

```python
states.groupby('State_name')['Alliance'].count()
```

```python
states.groupby('State_name')['Alliance'].value_counts().unstack()
```

```python
states.groupby('State_name')['Alliance'].value_counts().unstack().plot(kind='bar', stacked=True, figsize=(10,6))
```
