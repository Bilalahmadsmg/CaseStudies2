```python
electors2009.head()
```

```python
gender_counts = pd.DataFrame(candidates2009['Candidate_Sex'].value_counts())
gender_counts
```

```python
fig = plt.figure(figsize=(8,4))
plt.bar(height = gender_counts['Candidate_Sex'], x = gender_counts.index)
for num, val in enumerate (gender_counts['Candidate_Sex'].values):
    plt.text(x=num, y=val + 500, s=str(val),color='r')
plt.xlabel('Candidate Sex')
plt.ylabel('Count')
plt.ylim(0,9000)
plt.title('Candidate Female and Male')
plt.show()
```

#### Task: Create similar plots for each state

```python
sns.catplot(x='Candidate_Sex', col ='State_name', kind ='count', data=candidates2009, col_wrap=3, sharey=False )
plt.show()
```

#### Task: Gender plot for Maha and Goa bifurcated by parties

```python
sns.catplot(x='Candidate_Sex',col ='State_name', kind='count', data=candidates2009[candidates2009['State_name'].isin(['Maharashtra','Goa'])], hue = 'Party_Abbreviation', col_wrap=1, sharey=False, sharex=False)
```

```python
sns.catplot(x='Candidate_Sex',col ='State_name', kind='count', data=candidates2009[candidates2009['State_name'].isin(['Maharashtra','Goa']) & candidates2009['Party_Abbreviation'].isin(['BJP','INC', 'IND'])], hue = 'Party_Abbreviation', col_wrap=2, sharey=False, sharex=False)
plt.show()
```
